```python
for index, row in tqdm(test_df.iterrows())   :
      #===================================FIX URGENT
      #TODO: 
      # There is a mistake in label segmentation
      # hateful = 1 and non-hateful = 0
      # it is opposite NOW!

      #TRAIN AND TEST SET SHOULD BE MERGED BECUASE WE HAVE OTHER TRAINING SET !

      #==================================== 
      if row['label'] == 1:
         label = "hateful"
      else:
        label = "non-hateful"
      #label = "hateful" if row['label'] == 1 else "non-hateful"
      
      path_in = img_in + row['img']
      #print(path_in)
      img_path = row['img'].replace("img/","")
      path_out = img_out + label + "/" + img_path


      # Check if the input image exist
      if not os.path.isfile(path_in):
          continue

      # Check if the output dir of the label exist
      if not os.path.isdir(img_out + label):
          ss = img_out + label
          os.mkdir(ss)
          #os.mkdir(ss+"/"+"img")
          print("Directory for the label " + label + " created!")

      # Copy the image
      copyfile(path_in, path_out)
```

```python
from hugsvision.dataio.VisionDataset import VisionDataset

_, test, id2label, label2id = VisionDataset.fromImageFolder(
	"/content/small_test_data_set",
	test_ratio   = 1.00,
	balanced     = False,
	augmentation = False,
)
```

```python
test_df.label.value_counts()
```

```python
label2id['hateful'] = '1'
label2id['non-hateful'] = '0'
```

```python
label2id
```

```python
id2label
id2label['0'] = 'non-hateful'
id2label['1'] = 'hateful'
id2label
```

```python
huggingface_model = 'google/vit-base-patch16-224-in21k'
```

```python
from transformers import ViTFeatureExtractor, ViTForImageClassification
from hugsvision.nnet.VisionClassifierTrainer import VisionClassifierTrainer

trainer = VisionClassifierTrainer(
	model_name   = "HateClassifier",
	train      	 = train,
	test      	 = test,
	output_dir   = "./out/",
	max_epochs   = 7,
	batch_size   = 16, # On RTX 2080 Ti
	model = ViTForImageClassification.from_pretrained(
	    huggingface_model,
	    num_labels = len(label2id),
	    label2id   = label2id,
	    id2label   = id2label
	),
	feature_extractor = ViTFeatureExtractor.from_pretrained(
		huggingface_model,
	),
)
```

```python
ref, hyp = trainer.evaluate_f1_score()
```
